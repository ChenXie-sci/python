```python
# calculation of Cl,B atom's polarizability
import netCDF4
import os
import numpy as np
Lx = 59.39
L2x = 59.39/2
Ly = 60.018 
L2y = 60.018/2
Lz = 74.345
L2z = 74.345/2
#periodic boundary condition
def pbcx(a):
    while((a > L2x) | (a <= -L2x)):
        if(a > L2x):
            a -= Lx
        if(a <= -L2x):
            a += Lx
    return(a)
def pbcy(a):
    while((a > L2y) | (a <= -L2y)):
        if(a > L2y):
            a -= Ly
        if(a <= -L2y):
            a += Ly
    return(a)
def pbcz(a):
    while((a > L2z) | (a <= -L2z)):
        if(a > L2z):
            a -= Lz
        if(a <= -L2z):
            a += Lz
    return(a)
# build function of polar
def tensor(x1,y1,z1,x2,y2,z2):
    x = pbcx(x1 - x2)
    y = pbcy(y1 - y2)
    z = pbcz(z1 - z2)
    r = np.sqrt(x*x + y*y + z*z)
    a = np.eye(3) /(r * r * r) - np.array([[x*x,x*y,x*z],[y*x,y*y,y*z],[z*x,z*y,z*z]]) * 3 /(r * r * r * r * r)
    return a
# get coordinates of B and Cl atoms    
nc = netCDF4.Dataset('prd_hollow_1.nc')
c=nc.variables['coordinates']
d=np.zeros([72,3])
e=np.zeros([72,3])
j=0
for i in range(1501,2557,44):
    d[j]=c[0,i]
    e[j]=c[0,i-1]
    j+=1
for i in range(4101,5157,44):
    d[j]=c[0,i]
    e[j]=c[0,i-1]
    j+=1
for i in range(6701,7757,44):
    d[j]=c[0,i]
    e[j]=c[0,i-1]
    j+=1
Xe = np.zeros(144)
Ye = np.zeros(144)
Ze = np.zeros(144)
for i in range(72):
    Xe[i] = e[i][0]
    Ye[i] = e[i][1]
    Ze[i] = e[i][2]
for i in range(72,144):
    Xe[i] = d[i-72][0]
    Ye[i] = d[i-72][1]
    Ze[i] = d[i-72][2]
#b is the polarizability of single Cl atom, c is the polariability f single B atom
b = np.eye(3) * 20.47
#bf = b^(-1) matrix
bf = np.linalg.inv(b)
c = np.eye(3) * 14.71
#cf = c^(-1) matrix
cf = np.linalg.inv(c)
f = np.empty((144,144),dtype=object)
for i in range(72):
    for j in range(72):
        if(i == j):
            f[i][j] = cf
for i in range(72,144):
    for j in range(72,144):
        if(i == j):
            f[i][j] = bf
for i in range(144):
    for j in range(144):
        if(i != j):
            f[i][j] = tensor(Xe[i],Ye[i],Ze[i],Xe[j],Ye[j],Ze[j])
#calculate the B matrix
g = np.empty((1,144),dtype=object)
for i in range(144):
    g[0][i] = f[0][i]
for i in range(144):
    for j in range(1,144):
        g[0][i] = np.vstack((g[0][i],f[j][i]))
for i in range(1,144):
    g[0][0] = np.hstack((g[0][0],g[0][i]))
B = np.linalg.inv(g[0][0])
# Partition B matrix
row = np.hsplit(B,144)
C = np.empty(144,dtype=object)
for i in range(144):
        C[i] = np.vsplit(row[i],144)
# calculate the sum of the B_ij matrix
sum = np.zeros([3,3])
for i in range(144):
    for j in range(144):
        sum += C[i][j]
print(sum)            
```
