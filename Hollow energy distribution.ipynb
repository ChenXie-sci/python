```python
import matplotlib.pyplot as plt
import numpy as np
a=np.loadtxt('ex-1.txt')
b=np.loadtxt('ct-1.txt')
c=0.04336341046*(a-b)-0.10393
d=np.zeros(1000)
e=np.arange(-1.2,-0.58,0.0009809809809)
for i in range(2160000):
    for j in range(999):
        if((c[i] > e[j]) & (c[i] < e[j+1])):
            d[j]+=1
f=d/(0.0009809809809*2160000)
plt.title('Hollow energy gap distribution of 42 trajectories')
plt.xlabel('U(EX1-CT1)(eV)')
plt.ylabel('Probability Density')
plt.plot(e,f,color='black',label='Hollow distribution')
plt.legend()
plt.savefig('4.png')
plt.show()
```
